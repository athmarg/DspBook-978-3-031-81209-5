```python
import sympy as sp

t = sp.symbols('t', real=True)
u = sp.Heaviside

# Define impulse responses
h_a = sp.exp(-4*t) * u(t-2)
h_b = sp.exp(-6*t) * u(3-t)
h_c = sp.exp(-2*t) * u(t+50)

# Compute L1 norms
int_a = sp.integrate(sp.Abs(h_a), (t, -sp.oo, sp.oo))
int_b = sp.integrate(sp.Abs(h_b), (t, -sp.oo, sp.oo))
int_c = sp.integrate(sp.Abs(h_c), (t, -sp.oo, sp.oo))

print("Integral |h_a(t)| dt =", int_a)
print("Integral |h_b(t)| dt =", int_b)
print("Integral |h_c(t)| dt =", int_c)

# Check BIBO stability
def check_stable(val):
    return "Stable" if val.is_finite else "Unstable"

print("System a is", check_stable(int_a))
print("System b is", check_stable(int_b))
print("System c is", check_stable(int_c))
```
