```python
import sympy as sp

t = sp.symbols('t', real=True)
u = sp.Heaviside

# Define impulse responses
h_alpha = sp.exp(2*t) * u(-1-t)          # h(t) = e^{2t} u(-1-t)
h_beta  = sp.exp(-6*sp.Abs(t))           # h(t) = e^{-6|t|}
h_gamma = t * sp.exp(-t) * u(t)          # h(t) = t e^{-t} u(t)

# Compute L1 norms (integral of |h(t)| over all t)
int_alpha = sp.integrate(sp.Abs(h_alpha), (t, -sp.oo, sp.oo))
int_beta  = sp.integrate(sp.Abs(h_beta),  (t, -sp.oo, sp.oo))
int_gamma = sp.integrate(sp.Abs(h_gamma), (t, -sp.oo, sp.oo))

print("Integral |h_alpha(t)| dt =", int_alpha)
print("Integral |h_beta(t)| dt  =", int_beta)
print("Integral |h_gamma(t)| dt =", int_gamma)

# Check BIBO stability
def check_stable(val):
    return "Stable" if val.is_finite else "Unstable"

print("System α is", check_stable(int_alpha))
print("System β is", check_stable(int_beta))
print("System γ is", check_stable(int_gamma))
```
