```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from scipy.signal import convolve

# time axis
t = np.linspace(-1, 8, 2000)
dt = t[1] - t[0]
xi = 2

def u(t):
    return np.where(t >= 0, 1, 0)

def plot_system(alpha, beta, A):

    # signals
    x = A * (u(t) - u(t-xi))
    h = alpha * np.exp(-beta*t) * u(t)

    # analytical output
    y = (alpha*A/beta)*((1-np.exp(-beta*t))*u(t)
                        -(1-np.exp(-beta*(t-xi)))*u(t-xi))

    # numerical convolution
    y_num = convolve(x,h,mode='full') * dt
    t_conv = np.linspace(2*t[0],2*t[-1],len(y_num))

    # plots
    fig, ax = plt.subplots(1,3,figsize=(15,4))

    # x(t)
    ax[0].plot(t,x)
    ax[0].set_title("Input $x(t)$")
    ax[0].set_xlim(-1,8)
    ax[0].set_ylim(-0.5,5)
    ax[0].grid()

    # h(t)
    ax[1].plot(t,h)
    ax[1].set_title("Impulse response $h(t)$")
    ax[1].set_xlim(-1,8)
    ax[1].set_ylim(-0.5,5)
    ax[1].grid()

    # y(t)
    ax[2].plot(t,y,label="Analytical")
    ax[2].plot(t_conv,y_num,'--',label="Convolution")
    ax[2].set_xlim(-1,8)
    ax[2].set_ylim(-0.5,10)
    ax[2].set_title("Output $y(t)$")
    ax[2].legend()
    ax[2].grid()

    plt.tight_layout()
    plt.show()


interact(
    plot_system,
    alpha=FloatSlider(min=0.1,max=5,step=0.1,value=2,description='α'),
    beta=FloatSlider(min=0.1,max=5,step=0.1,value=1,description='β'),
    A=FloatSlider(min=0.1,max=5,step=0.1,value=1,description='A')
);
```
