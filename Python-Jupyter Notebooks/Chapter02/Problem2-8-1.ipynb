```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 1000)
dt = t[1] - t[0]

x = np.sin(t)

# ------------------------
# (α) Time delay
# ------------------------
delay = 6
y_a = np.interp(t, t - delay, x, left=0, right=0)
x_rec_a = np.interp(t, t + delay, y_a, left=0, right=0)

plt.figure(figsize=(10,3))
plt.plot(t, x, color='blue', linewidth=1.5, label='x(t)')
plt.plot(t, x_rec_a, color='red', linestyle=(0,(10,8)), linewidth=1.5, label='Recovered x(t)')
plt.plot(t, y_a, color='green', alpha=0.7, label='y(t) = x(t-6)')

# Διακεκομμένες μαύρες γραμμές για το μη μηδενικό μέρος
plt.vlines([delay, 10], ymin=-1, ymax=1, colors='black', linestyles='dashed', linewidth=1, label='Non-zero region')

plt.title(r'System (α): $y(t) = x(t-6)$ and its inverse')
plt.legend(loc='lower left', bbox_to_anchor=(0.1, 0.01))
plt.show()

# ------------------------
# (β) Integrator
# ------------------------
y_b = np.cumsum(x)*dt
x_rec_b = np.gradient(y_b, dt)

plt.figure(figsize=(10,3))
plt.plot(t, x, color='blue', linewidth=1.5, label='x(t)')
plt.plot(t, x_rec_b, color='red', linestyle=(0,(10,8)), linewidth=1.5, label='Recovered x(t)')
plt.plot(t, y_b, color='green', alpha=0.7, label=r'$y(t) = \int x(\tau) d\tau$')

plt.title(r'System (β): $y(t) = \int x(\tau)d\tau$ and its inverse')
plt.legend(loc='lower left', bbox_to_anchor=(0.65, 0.01))
plt.show()

# ------------------------
# (γ) Exponential kernel
# ------------------------
h = np.exp(-t)
y_c = np.convolve(x, h)[:len(t)]*dt
x_rec_c = np.gradient(y_c, dt) + y_c

plt.figure(figsize=(10,3))
plt.plot(t, x, color='blue', linewidth=1.5, label='x(t)')
plt.plot(t, x_rec_c, color='red', linestyle=(0,(10,8)), linewidth=1.5, label='Recovered x(t)')
plt.plot(t, y_c, color='green', alpha=0.7, label=r'$y(t) = \int_0^t e^{-(t-\tau)} x(\tau)d\tau$')

plt.title(r'System (γ): $y(t) = \int_0^t e^{-(t-\tau)} x(\tau)d\tau$ and its inverse')
plt.legend()
plt.show()
```
