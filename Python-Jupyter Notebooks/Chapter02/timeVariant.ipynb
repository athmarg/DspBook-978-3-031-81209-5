```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# time axis
t = np.linspace(-5,10,2000)

# input signal
x = np.exp(-0.2*t)*np.sin(2*t)

def plot_systems(tau):

    # shifted input
    x_shift = np.exp(-0.2*(t - tau))*np.sin(2*(t - tau))

    # ---------------------
    # System 1: y(t)=x(t-6)  (time invariant)

    y_shift = np.exp(-0.2*((t - tau) - 6))*np.sin(2*((t - tau) - 6))
    Tx_shift = np.exp(-0.2*((t - 6) - tau))*np.sin(2*((t - 6) - tau))

    # ---------------------
    # System 2: y(t)=x(t)cos(t) (time varying)

    y2_shift = x_shift * np.cos(t - tau)   # y(t-τ)
    Tx2_shift = x_shift * np.cos(t)        # T[x(t-τ)]

    # ---------------------
    # Plot

    fig,axs = plt.subplots(1,2,figsize=(12,4))

    # time invariant
    axs[0].plot(t,y_shift,label='y(t-τ)')
    axs[0].plot(t,Tx_shift,'--',label='T[x(t-τ)]')
    axs[0].set_title('Time-invariant system: y(t)=x(t-6)')
    axs[0].grid(True)
    axs[0].legend()

    # time varying
    axs[1].plot(t,y2_shift,label='y(t-τ)')
    axs[1].plot(t,Tx2_shift,'--',label='T[x(t-τ)]')
    axs[1].set_title('Time-varying system: y(t)=x(t)cos(t)')
    axs[1].grid(True)
    axs[1].legend()

    plt.show()


# slider
interact(
    plot_systems,
    tau=FloatSlider(value=0,min=-5,max=5,step=0.2,description='τ')
)
```
