```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Layout

# Time axis
t = np.linspace(0, 20, 1000)
dt = t[1]-t[0]

# ------------------------
# Stable system
def stable_system(x):
    return np.sin(x)  # BIBO stable

# Inputs for stable system
xs1 = np.exp(-0.2*t) * np.sin(2*t)
xs2 = np.cos(t)

# Outputs for stable system
ys1 = stable_system(xs1)
ys2 = stable_system(xs2)

# ------------------------
# Unstable system
def unstable_system(t, x):
    return t * x

# Inputs for unstable system
xu1 = np.exp(-0.2*t) * np.sin(2*t)  # appears bounded
xu2 = np.ones_like(t)                # clearly unbounded

# Outputs for unstable system
yu1 = unstable_system(t, xu1)
yu2 = unstable_system(t, xu2)

# ------------------------
# Slider plot function
def plot_stability(time_idx):
    idx = int(time_idx)
    
    fig, axs = plt.subplots(2,3, figsize=(18,8))
    plt.subplots_adjust(wspace=0.3, hspace=0.4, bottom=0.2)
    
    # ---- Stable system (top row)
    axs[0,0].plot(t[:idx], xs1[:idx], 'b', lw=2)
    axs[0,0].plot(t[:idx], ys1[:idx], 'r', lw=2)
    axs[0,0].set_title('Stable system: Input x1 and output y(t)')
    axs[0,0].grid(True)
    axs[0,0].text(0.5, -0.2, 'Blue: x1(t), Red: y(t) = sin(x1(t))', 
                  transform=axs[0,0].transAxes, ha='center', fontsize=10)
    
    axs[0,1].plot(t[:idx], xs2[:idx], 'b', lw=2)
    axs[0,1].plot(t[:idx], ys2[:idx], 'r', lw=2)
    axs[0,1].set_title('Stable system: Input x2 and output y(t)')
    axs[0,1].grid(True)
    axs[0,1].text(0.5, -0.2, 'Blue: x2(t), Red: y(t) = sin(x2(t))', 
                  transform=axs[0,1].transAxes, ha='center', fontsize=10)
    
    axs[0,2].plot(t[:idx], ys1[:idx], 'g', lw=2)
    axs[0,2].plot(t[:idx], ys2[:idx], 'm', lw=2)
    axs[0,2].set_title('Stable system: Compare outputs')
    axs[0,2].grid(True)
    axs[0,2].text(0.5, -0.2, 'Green: y(t) for x1(t), Magenta: y(t) for x2(t)', 
                  transform=axs[0,2].transAxes, ha='center', fontsize=10)
    
    # ---- Unstable system (bottom row)
    axs[1,0].plot(t[:idx], xu1[:idx], 'b', lw=2)
    axs[1,0].plot(t[:idx], yu1[:idx], 'r', lw=2)
    axs[1,0].set_title('Unstable system: Input x1 and output y(t)')
    axs[1,0].grid(True)
    axs[1,0].text(0.5, -0.2, 'Blue: x1(t), Red: y(t) = t*x1(t)', 
                  transform=axs[1,0].transAxes, ha='center', fontsize=10)
    
    axs[1,1].plot(t[:idx], xu2[:idx], 'b', lw=2)
    axs[1,1].plot(t[:idx], yu2[:idx], 'r', lw=2)
    axs[1,1].set_title('Unstable system: Input x2 and output y(t)')
    axs[1,1].grid(True)
    axs[1,1].text(0.5, -0.2, 'Blue: x2(t), Red: y(t) = t*x2(t)', 
                  transform=axs[1,1].transAxes, ha='center', fontsize=10)
    
    axs[1,2].plot(t[:idx], yu1[:idx], 'g', lw=2)
    axs[1,2].plot(t[:idx], yu2[:idx], 'm', lw=2)
    axs[1,2].set_title('Unstable system: Compare outputs')
    axs[1,2].grid(True)
    axs[1,2].text(0.5, -0.2, 'Green: y(t) for x1(t), Magenta: y(t) for x2(t)', 
                  transform=axs[1,2].transAxes, ha='center', fontsize=10)
    
    plt.show()

# Slider
time_slider = FloatSlider(value=8, min=8, max=len(t)-1, step=1, description='Time', layout=Layout(width='90%'))
interact(plot_stability, time_idx=time_slider)
```
