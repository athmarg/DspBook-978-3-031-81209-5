```python
# RC Circuit Responses Notebook
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Parameters
# -----------------------------
R = 1.0        # Resistance in ohms
C = 1.0        # Capacitance in farads
tau = R*C      # Time constant
omega = 2.0    # Frequency for sinusoidal input
t = np.linspace(0, 10, 1000)  # Time vector

# -----------------------------
# 1. Impulse Response h(t)
# h(t) = (1/RC) * exp(-t/RC) * u(t)
# -----------------------------
h = (1/(R*C)) * np.exp(-t/(R*C))

# Plot Impulse Response
plt.figure(figsize=(8,4))
plt.plot(t, h)
plt.title("Impulse Response h(t)")
plt.xlabel("Time t")
plt.ylabel("h(t)")
plt.grid(True)
plt.show()

# -----------------------------
# 2. Step Response s(t)
# s(t) = 1 - exp(-t/RC) * u(t)
# -----------------------------
s = 1 - np.exp(-t/(R*C))

# Plot Step Response
plt.figure(figsize=(8,4))
plt.plot(t, s)
plt.title("Step Response s(t)")
plt.xlabel("Time t")
plt.ylabel("s(t)")
plt.grid(True)
plt.show()

# -----------------------------
# 3. Sinusoidal Input x(t)
# x(t) = sin(ωt)
# -----------------------------
x = np.sin(omega*t)

# Plot Input Signal
plt.figure(figsize=(8,4))
plt.plot(t, x)
plt.title("Input Signal x(t) = sin(ωt)")
plt.xlabel("Time t")
plt.ylabel("x(t)")
plt.grid(True)
plt.show()

# -----------------------------
# 4. Output y(t) for Sinusoidal Input
# y(t) = [sin(ωt) + ωRC (exp(-t/RC) - cos(ωt))] / (1 + (ωRC)^2)
# -----------------------------
y = (np.sin(omega*t) + omega*R*C*(np.exp(-t/(R*C)) - np.cos(omega*t))) / (1 + (omega*R*C)**2)

# Plot Input vs Output
plt.figure(figsize=(8,4))
plt.plot(t, x, label="Input x(t)")
plt.plot(t, y, label="Output y(t)")
plt.title("Response of RC Circuit to Sinusoidal Input")
plt.xlabel("Time t")
plt.ylabel("Signal")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# 5. Magnitude of Response to Sinusoidal Input |H(ω)|
# |H(ω)| = 1 / sqrt(1 + (ωRC)^2)
# -----------------------------
omega_vals = np.linspace(0, 10, 500)
Hmag = 1/np.sqrt(1 + (omega_vals*R*C)**2)

# Plot Frequency Response Magnitude
plt.figure(figsize=(8,4))
plt.plot(omega_vals, Hmag)
plt.title("Magnitude of Response to Sinusoidal Input |H(ω)|")
plt.xlabel("ω (rad/s)")
plt.ylabel("|H(ω)|")
plt.grid(True)
plt.show()
```
