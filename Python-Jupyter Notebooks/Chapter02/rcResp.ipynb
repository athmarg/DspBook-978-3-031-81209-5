```python
# RC Circuit Responses - 2x2 Grid Plot
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Parameters
# -----------------------------
R = 1.0        # Resistance in ohms
C = 1.0        # Capacitance in farads
tau = R*C      # Time constant
omega = 2.0    # Frequency for sinusoidal input
t = np.linspace(0, 10, 1000)  # Time vector

# -----------------------------
# 1. Impulse Response h(t)
# -----------------------------
h = (1/(R*C)) * np.exp(-t/(R*C))

# -----------------------------
# 2. Step Response s(t)
# -----------------------------
s = 1 - np.exp(-t/(R*C))

# -----------------------------
# 3. Input Sinusoid x(t)
# -----------------------------
x = np.sin(omega*t)

# -----------------------------
# 4. Output y(t) for Sinusoidal Input
# y(t) = [sin(ωt) + ωRC (exp(-t/RC) - cos(ωt))] / (1 + (ωRC)^2)
# -----------------------------
y = (np.sin(omega*t) + omega*R*C*(np.exp(-t/(R*C)) - np.cos(omega*t))) / (1 + (omega*R*C)**2)

# -----------------------------
# Plot all 4 in a 2x2 grid
# -----------------------------
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Impulse Response
axs[0, 0].plot(t, h, color='tab:blue')
axs[0, 0].set_title("Impulse Response h(t)")
axs[0, 0].set_xlabel("t")
axs[0, 0].set_ylabel("h(t)")
axs[0, 0].grid(True)

# Step Response
axs[0, 1].plot(t, s, color='tab:orange')
axs[0, 1].set_title("Step Response s(t)")
axs[0, 1].set_xlabel("t")
axs[0, 1].set_ylabel("s(t)")
axs[0, 1].grid(True)

# Input Sinusoid
axs[1, 0].plot(t, x, color='tab:green')
axs[1, 0].set_title("Input Signal x(t) = sin(ωt)")
axs[1, 0].set_xlabel("t")
axs[1, 0].set_ylabel("x(t)")
axs[1, 0].grid(True)

# Output Sinusoid
axs[1, 1].plot(t, y, color='tab:red')
axs[1, 1].plot(t, x, '--', color='tab:gray', label="Input x(t)")
axs[1, 1].set_title("Output y(t) to Sinusoidal Input")
axs[1, 1].set_xlabel("t")
axs[1, 1].set_ylabel("y(t)")
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
