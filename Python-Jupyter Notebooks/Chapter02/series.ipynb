```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 500)
dt = t[1]-t[0]
x = np.exp(-0.2*t) * np.sin(2*np.pi*t)
y1 = np.cumsum(x)*dt
y = 0.5*y1 + 0.3*np.sin(0.5*t)

fig, ax = plt.subplots(figsize=(12,6))

systems = [ (2,4,'T1',r'$y_1(t)=\int x(t)d\tau$','lightblue'),
            (6,4,'T2',r'$y(t)=0.5y_1(t)+0.3\sin(0.5t)$','lightgreen') ]

print("x(t) = e^{-0.2t} sin(2πt)")

print("\ny1(t) = ∫_0^t e^{-0.2τ} sin(2π τ) dτ")
print("     = [2π - e^{-0.2t}(0.2 sin(2πt) + 2π cos(2πt))] / (0.04 + 4π²)")

print("\ny(t) = 0.5 y1(t) + 0.3 sin(0.5t)")
print("     = 0.5 * [2π - e^{-0.2t}(0.2 sin(2πt) + 2π cos(2πt))] / (0.04 + 4π²) + 0.3 sin(0.5t)")

for x0,y0,label,eq,color in systems:
    ax.add_patch(plt.Rectangle((x0,y0),2,1,facecolor=color,edgecolor='black'))
    ax.text(x0+1,y0+0.5,label,ha='center',va='center',fontsize=12)
    ax.text(x0+1,y0+1.2,eq,ha='center',va='center',fontsize=12)
arrows=[(0.5,4.5,2,4.5),(4,4.5,6,4.5),(8,4.5,10.5,4.5)]
for xs,ys,xe,ye in arrows:
    ax.annotate('',xy=(xe,ye),xytext=(xs,ys),
                arrowprops=dict(arrowstyle='->',lw=2))
ax.text(0.2,4.5,r'$x(t)$',ha='center',va='center',fontsize=12)
ax.text(10.8,4.5,r'$y(t)$',ha='center',va='center',fontsize=12)
signals = [ (2.8,x), (5.8,y1), (8.7,y) ]

for xc,sig in signals:
    width = 0.10
    height = 0.08
    left = xc/12 - width/2    
    bottom = 0.55             
    inset = fig.add_axes([left,bottom,width,height])
    inset.plot(t,sig,color='red',lw=1)
    inset.set_xlim(t[0],t[-1])
    inset.set_ylim(np.min(sig)*1.05,np.max(sig)*1.05)
    inset.set_xticks([])
    inset.set_yticks([])
    inset.axis('off')
ax.set_xlim(-1,12)
ax.set_ylim(0,6)
ax.axis('off')
plt.show()
```
