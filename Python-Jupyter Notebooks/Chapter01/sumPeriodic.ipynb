```python
# =============================================================================
# Demonstration: Periodicity of the sum of two periodic signals
# x(t) period 15 ms, y(t) period 20 ms → z(t) period 60 ms
# Suitable for Springer book / academic lecture notes
# =============================================================================

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# =============================================================================
# Basic Parameters (in seconds)
# =============================================================================
T1    = 15e-3           # περίοδος x(t) → 15 ms
T2    = 20e-3           # περίοδος y(t) → 20 ms
T_sum = 60e-3           # LCM(15 ms, 20 ms) = 60 ms

# Χρονικός άξονας — δείχνουμε λίγο παραπάνω από 2 περιόδους του αθροίσματος
t = np.linspace(0, 2.2 * T_sum, 8000)

# =============================================================================
# The main function call
# =============================================================================
def plot_signals(
    amp_x   = 1.0,
    amp_y   = 1.0,
    phase_x = 0.0,
    phase_y = 0.0
):
    # Create Signals
    x = amp_x * np.sin(2 * np.pi * t / T1 + phase_x)
    y = amp_y * np.sin(2 * np.pi * t / T2 + phase_y)
    z = x + y

    # Clear previous plot (if any)
    plt.clf()

    # Create new plot
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8.5),
                                        sharex=True,
                                        gridspec_kw={'height_ratios': [1, 1, 1.3]})

    plt.subplots_adjust(left=0.12, right=0.95, bottom=0.08, top=0.95, hspace=0.38)

    # x(t)
    ax1.plot(t*1000, x, 'b-', lw=1.3, label='x(t)')
    ax1.set_title('x(t)  –  fundamental period = 15 ms', fontsize=11)
    ax1.grid(True, alpha=0.3, linestyle='--')
    ax1.set_ylabel('Amplitude')
    ax1.legend(loc='upper right', fontsize=9)

    # y(t)
    ax2.plot(t*1000, y, 'r-', lw=1.3, label='y(t)')
    ax2.set_title('y(t)  –  fundamental period = 20 ms', fontsize=11)
    ax2.grid(True, alpha=0.3, linestyle='--')
    ax2.set_ylabel('Amplitude')
    ax2.legend(loc='upper right', fontsize=9)

    # z(t) = x + y
    ax3.plot(t*1000, z, 'k-', lw=1.7, label='z(t) = x(t) + y(t)')
    ax3.set_title('z(t)  –  fundamental period = 60 ms (LCM(15,20) ms)', fontsize=11)
    ax3.grid(True, alpha=0.3, linestyle='--')
    ax3.set_ylabel('Amplitude')
    ax3.set_xlabel('Time (ms)')
    ax3.legend(loc='upper right', fontsize=9)

    # Κατακόρυφες γραμμές κάθε 60 ms
    for k in [1, 2]:
        ax3.axvline(x=k*60, color='gray', ls='--', lw=1.1, alpha=0.55)

    plt.show()


# =============================================================================
# Δημιουργία sliders
# =============================================================================
interact(
    plot_signals,
    amp_x   = FloatSlider(value=1.0, min=0.3, max=2.0, step=0.1, description='Amp. x(t):', readout_format='.1f'),
    amp_y   = FloatSlider(value=1.0, min=0.3, max=2.0, step=0.1, description='Amp. y(t):', readout_format='.1f'),
    phase_x = FloatSlider(value=0.0, min=-np.pi, max=np.pi, step=0.1, description='Ph. x(t) [rad]:', readout_format='.1f'),
    phase_y = FloatSlider(value=0.0, min=-np.pi, max=np.pi, step=0.1, description='Ph. y(t) [rad]:', readout_format='.1f'),
    continuous_update=True
);
```
