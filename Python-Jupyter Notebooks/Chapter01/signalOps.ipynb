```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, RadioButtons

t = np.linspace(-1, 1, 1000)

def x_signal(t, A1, f1):
    return A1 * np.sin(2*np.pi*f1*t)

def y_signal(t, A2, f2):
    return A2 * np.cos(2*np.pi*f2*t)

def plot_signals(operation, A1, f1, A2, f2):

    x = x_signal(t, A1, f1)
    y = y_signal(t, A2, f2)

    if operation == 'Addition':
        result = x + y
        title = "x(t) + y(t)"

    elif operation == 'Multiplication':
        result = x * y
        title = "x(t) · y(t)"

    elif operation == 'Composition x(y(t))':
        result = x_signal(y, A1, f1)
        title = "x(y(t))"

    fig, ax = plt.subplots(3,1, figsize=(8,8))

    ax[0].plot(t, x)
    ax[0].set_title("Signal x(t)")
    ax[0].grid(True)

    ax[1].plot(t, y)
    ax[1].set_title("Signal y(t)")
    ax[1].grid(True)

    ax[2].plot(t, result, 'r')
    ax[2].set_title(f"Result: {title}")
    ax[2].grid(True)
    ax[2].set_xlabel("Time")

    plt.tight_layout()
    plt.show()


interact(
    plot_signals,
    
    operation=RadioButtons(
        options=[
            'Addition',
            'Multiplication',
            'Composition x(y(t))'
        ],
        value='Addition',   # default επιλογή
        description='Operation:'
    ),
    
    A1=FloatSlider(min=0.5, max=2, step=0.1, value=1, description='A1'),
    f1=FloatSlider(min=0.5, max=5, step=0.5, value=1, description='f1'),
    
    A2=FloatSlider(min=0.5, max=2, step=0.1, value=1, description='A2'),
    f2=FloatSlider(min=0.5, max=5, step=0.5, value=2, description='f2'),
);
```
