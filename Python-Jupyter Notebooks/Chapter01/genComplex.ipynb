```python
"""
Interactive General Complex Exponential Signal
Author: Athanasios I. Margaris

Signal:
    x(t) = C * exp(alpha * t) = |C| e^{r t} e^{j(ω0 t + θ)}

Features:
- Sliders for |C|, r (real part of alpha), ω0 (imag part), θ (phase of C)
- Complex plane phasor visualization (spiral if r≠0)
- Real and Imaginary parts vs time
- Envelopes ±|C| e^{r t}
"""

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def complex_general_demo(C_mag=1.0, r=0.0, omega0=2.0, theta=0.0, t_slider=0.0):
    
    # Time axis for signal
    t_axis = np.linspace(0, 10, 1000)
    
    # General complex exponential
    x = C_mag * np.exp(r * t_axis) * np.exp(1j*(omega0*t_axis + theta))
    
    real_part = np.real(x)
    imag_part = np.imag(x)
    
    # Current phasor at t_slider
    xt = C_mag * np.exp(r*t_slider) * np.exp(1j*(omega0*t_slider + theta))
    xr = np.real(xt)
    xi = np.imag(xt)
    
    envelope = C_mag * np.exp(r * t_axis)
    
    fig, ax = plt.subplots(1,2, figsize=(14,5))
    
    # -------- Complex Plane (Phasor / Spiral) --------
    ax[0].set_title("Complex Plane (Phasor / Spiral)")
    ax[0].plot(real_part, imag_part, color='lightgray', label='Path')
    ax[0].plot([0,xr],[0,xi], linewidth=3, color='blue', label='Phasor')
    ax[0].scatter(xr, xi, color='red')
    
    # optional circle at t=0 for reference
    circle = plt.Circle((0,0),C_mag,fill=False,linestyle='--', color='gray')
    ax[0].add_patch(circle)
    
    ax[0].set_xlabel("Real")
    ax[0].set_ylabel("Imaginary")
    ax[0].grid(True)
    ax[0].set_aspect('equal')
    
    # -------- Real / Imag Components vs Time --------
    ax[1].set_title("Real and Imaginary Parts vs Time")
    ax[1].plot(t_axis, real_part, label='Real Part', color='blue')
    ax[1].plot(t_axis, imag_part, label='Imag Part', color='orange')
    
    # Plot envelopes
    ax[1].plot(t_axis, envelope, 'k--', label='+Envelope')
    ax[1].plot(t_axis, -envelope, 'k--', label='-Envelope')
    
    # Show current t_slider
    ax[1].axvline(t_slider, color='red', linestyle='--')
    ax[1].scatter(t_slider, xr, color='blue')
    ax[1].scatter(t_slider, xi, color='orange')
    
    ax[1].set_xlabel("Time t")
    ax[1].set_ylabel("Amplitude")
    ax[1].grid(True)
    ax[1].legend()
    
    plt.show()

# Interactive sliders
interact(complex_general_demo,
         C_mag=FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='|C|'),
         r=FloatSlider(value=0.0, min=-1.0, max=1.0, step=0.05, description='r'),
         omega0=FloatSlider(value=2.0, min=-10.0, max=10.0, step=0.2, description='ω0'),
         theta=FloatSlider(value=0.0, min=0.0, max=2*np.pi, step=0.1, description='θ'),
         t_slider=FloatSlider(value=0.0, min=0.0, max=10.0, step=0.05, description='t'));
```
