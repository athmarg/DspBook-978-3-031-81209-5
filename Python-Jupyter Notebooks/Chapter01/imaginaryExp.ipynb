```python
"""
Interactive Imaginary Exponential Signal
Author: Athanasios I. Margaris

Signal:
    x(t) = C * exp(j * omega * t)

Features:
- Slider for amplitude C
- Slider for angular frequency omega
- Displays:
    Real part
    Imaginary part
    Magnitude
"""

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider


def plot_complex_exponential(C=1.0, omega=2.0):

    t = np.linspace(0, 2*np.pi, 1000)

    x = C * np.exp(1j * omega * t)

    real_part = np.real(x)
    imag_part = np.imag(x)
    magnitude = np.abs(x)

    plt.figure(figsize=(10,5))

    plt.plot(t, real_part, label="Real Part (cos)", linewidth=2)
    plt.plot(t, imag_part, label="Imag Part (sin)", linewidth=2)
    plt.plot(t, magnitude, '--', label="Magnitude |x(t)|")

    plt.xlabel("Time t")
    plt.ylabel("Signal value")
    plt.title(f"Complex Exponential Signal: x(t)= {C:.2f} e^(j {omega:.2f} t)")
    plt.grid(True)
    plt.legend()

    plt.show()


interact(plot_complex_exponential,
         C=FloatSlider(value=1.0, min=0.0, max=5.0, step=0.5, description='C'),
         omega=FloatSlider(value=2.0, min=-10.0, max=10.0, step=0.5, description='omega'));
```
