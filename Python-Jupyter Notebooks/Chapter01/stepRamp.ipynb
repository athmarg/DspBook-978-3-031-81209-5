```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def plot_step_and_ramp_shift(A=1.0, t0=0.0):
    t = np.linspace(-5, 15, 1000)
    
    # Shifted unit step
    u = np.zeros_like(t)
    u[t >= t0] = 1
    step_signal = A * u
    
    # Shifted unit ramp
    ramp_signal = (t - t0) * u
    
    fig, ax = plt.subplots(1, 2, figsize=(14,5))
    
    # -------- Unit Step --------
    ax[0].plot(t, step_signal, color='blue', linewidth=2)
    ax[0].set_title(f'Unit Step Function: A*u(t-t0), A={A}, t0={t0}')
    ax[0].set_xlabel('t')
    ax[0].set_ylabel('Amplitude')
    ax[0].grid(True)
    ax[0].set_ylim(-0.5, 1.2*A)
    
    # -------- Unit Ramp --------
    ax[1].plot(t, ramp_signal, color='orange', linewidth=2)
    ax[1].set_title(f'Unit Ramp Function: r(t-t0)')
    ax[1].set_xlabel('t')
    ax[1].set_ylabel('Amplitude')
    ax[1].grid(True)
    ax[1].set_ylim(-1, np.max(ramp_signal)+1)
    
    plt.tight_layout()
    plt.show()

interact(plot_step_and_ramp_shift,
         A=FloatSlider(value=1.0, min=0.0, max=10.0, step=0.1, description='A'),
         t0=FloatSlider(value=0.0, min=0.0, max=10.0, step=0.1, description='t0'));
```
