```python
"""
Interactive Sine Wave
Author: Athanasios I. Margaris
Purpose: Demonstrate the effect of amplitude, frequency, and phase on a sine wave

Features:
- Plots the sine wave: y = A * sin(2π f t + φ)
- Interactive sliders for:
    - Amplitude (A)
    - Frequency (f)
    - Phase (φ)
- Real-time update of the waveform as sliders are moved
- Helps students visually understand the impact of each parameter

How to Use:
1. Open in JupyterLab, Jupyter Notebook, or Binder
2. Use the sliders to change the parameters
3. Observe waveform changes interactively

Requirements:
- Python 3.x
- numpy
- matplotlib
- ipywidgets
"""

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Function to plot sine wave
def plot_sine(amplitude=1.0, frequency=1.0, phase=0.0):
    t = np.linspace(0, 2, 1000)  # Time vector from 0 to 2 seconds
    y = amplitude * np.sin(2 * np.pi * frequency * t + phase)
    
    plt.figure(figsize=(8,4))
    plt.plot(t, y, label=f'Amplitude={amplitude}, Frequency={frequency} Hz, Phase={phase} rad')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title('Interactive Sine Wave')
    plt.grid(True)
    plt.legend()
    plt.ylim(-2, 2)
    plt.show()

# Interactive sliders
interact(plot_sine,
         amplitude=FloatSlider(value=1.0, min=0.0, max=2.0, step=0.1, description='Amplitude'),
         frequency=FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Frequency [Hz]'),
         phase=FloatSlider(value=0.0, min=0.0, max=2*np.pi, step=0.1, description='Phase [rad]'));

```
