```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def triangular_pulse(alpha=0.5):
    t = np.linspace(-2, 2, 1000)
    pulse = np.zeros_like(t)
    
    # Triangular pulse: 1 - |t|/alpha for |t| < alpha, 0 otherwise
    mask = np.abs(t) < alpha
    pulse[mask] = 1 - np.abs(t[mask])/alpha
    
    fig, ax = plt.subplots(figsize=(8,4))
    ax.plot(t, pulse, color='green', linewidth=2)
    ax.fill_between(t, 0, pulse, color='green', alpha=0.3)
    
    ax.set_title(f'Triangular Pulse Λ_α(t) with α={alpha}')
    ax.set_xlabel('t')
    ax.set_ylabel('Amplitude')
    ax.grid(True)
    ax.set_ylim(-0.1, 1.2)
    
    # Show base width and peak height
    width = 2*alpha
    height = 1
    ax.text(0.5, -0.15, f'Base Width = {width:.3f}, Peak Height = {height}',
            transform=ax.transAxes, ha='center', va='top', fontsize=12, color='black')
    
    plt.show()

# Slider for alpha
interact(triangular_pulse, alpha=FloatSlider(value=0.5, min=0.1, max=1.5, step=0.05, description='α'));
```
