```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatLogSlider

def delta_approx(epsilon=0.1):
    t = np.linspace(-1, 1, 1000)
    delta = np.zeros_like(t)
    
    # Rectangle from -ε/2 to ε/2 with height 1/ε
    delta[np.abs(t) <= epsilon/2] = 1/epsilon
    
    width = epsilon
    height = 1/epsilon
    area = width * height  # Should be ~1
    
    fig, ax = plt.subplots(figsize=(8,4))
    ax.plot(t, delta, color='red', linewidth=2)
    ax.fill_between(t, 0, delta, color='red', alpha=0.3)
    
    ax.set_title(f'Approximation of δ(t) with ε={epsilon}')
    ax.set_xlabel('t')
    ax.set_ylabel('Amplitude')
    ax.grid(True)
    
    # Annotation with width, height and area
    annotation_text = f'Area under curve = width × height = {width:.3f} × {height:.3f} ≈ {area:.3f}'
    ax.text(0.5, -0.15, annotation_text,
            transform=ax.transAxes, ha='center', va='top', fontsize=12, color='black')
    
    # Adjust y-limits a bit to make space for annotation
    ax.set_ylim(-0.2, max(delta)*1.1)
    
    plt.show()

# Logarithmic slider to see ε from 1 → 0.01
interact(delta_approx,
         epsilon=FloatLogSlider(value=0.1, base=10, min=-2, max=0, step=0.1, description='ε'));
```
