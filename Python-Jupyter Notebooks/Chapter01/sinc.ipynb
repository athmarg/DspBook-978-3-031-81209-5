```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def sinc_function(scale=1.0):
    t = np.linspace(-10, 10, 2000)
    # Normalized sinc: sin(scale*t)/(scale*t)
    y = np.sinc(scale*t/np.pi)
    
    fig, ax = plt.subplots(figsize=(8,4))
    ax.plot(t, y, color='purple', linewidth=2)
    ax.fill_between(t, 0, y, color='purple', alpha=0.2)
    
    ax.set_title(f'Sinc Function sinc(scale*t) with scale={scale}')
    ax.set_xlabel('t')
    ax.set_ylabel('Amplitude')
    ax.grid(True)
    ax.set_ylim(-0.25, 1.1)
    
    plt.show()

# Slider for scale factor
interact(sinc_function, scale=FloatSlider(value=1.0, min=0.2, max=5.0, step=0.1, description='Scale'));
```
