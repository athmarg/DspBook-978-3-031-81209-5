```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Checkbox

# Time axis
t = np.linspace(-10, 10, 2000)

# Example signal
def x(t):
    """
    Example continuous-time signal.
    A Gaussian-modulated cosine is used so that
    time transformations are clearly visible.
    """
    return np.exp(-0.2 * t**2) * np.cos(2*np.pi*0.5*t)


def plot_transform(alpha, T, reverse):
    """
    Demonstration of basic time transformations.

    alpha  : time scaling factor
    T      : time shift
    reverse: enables time reversal

    Transformed signal:

        y(t) = x(alpha*(t - T))

    If reversal is enabled:

        y(t) = x(alpha*(-t - T))
    """

    # Original signal
    x_original = x(t)

    # Apply transformations
    if reverse:
        y = x(alpha * (-t - T))
    else:
        y = x(alpha * (t - T))

    # Plot
    plt.figure(figsize=(10,5))

    plt.plot(t, x_original, label="Original signal x(t)", linewidth=2)
    plt.plot(t, y, label="Transformed signal y(t)", linewidth=2)

    plt.axhline(0)
    plt.axvline(0)

    plt.xlabel("Time t")
    plt.ylabel("Amplitude")
    plt.title("Continuous-Time Signal Transformations")

    plt.grid(True)
    plt.legend()
    plt.show()


interact(
    plot_transform,
    alpha=FloatSlider(value=1, min=0.2, max=3, step=0.1, description="Scaling α"),
    T=FloatSlider(value=0, min=-5, max=5, step=0.1, description="Shift T"),
    reverse=Checkbox(value=False, description="Time reversal")
);
```
