```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

def comb_function(T=1.0, N=5):
    """
    Plot the comb function:
    - T: period between impulses
    - N: number of impulses on each side of 0
    """
    
    # Positions of impulses
    n_values = np.arange(-N, N+1)  # from -N to N
    x_positions = n_values * T
    y_values = np.ones_like(x_positions)
    
    # Plot
    fig, ax = plt.subplots(figsize=(8,4))
    
    # stem without use_line_collection to avoid Jupyter errors
    markerline, stemlines, baseline = ax.stem(x_positions, y_values, linefmt='blue', markerfmt='bo', basefmt=' ')
    
    ax.set_title(f'Comb Function with period T={T:.2f} and N={N}')
    ax.set_xlabel('t')
    ax.set_ylabel('Amplitude')
    ax.grid(True)
    ax.set_ylim(0, 1.2)
    
    # Annotation for displayed range
    width = 2*N*T
    ax.text(0.5, -0.15, f'Displayed Range = ±{N*T:.2f}, Total Width = {width:.2f}', 
            transform=ax.transAxes, ha='center', va='top', fontsize=12, color='black')
    
    plt.show()

# Interactive sliders for T and N
interact(comb_function, 
         T=FloatSlider(value=1.0, min=0.1, max=5.0, step=0.05, description='T'),
         N=IntSlider(value=5, min=0, max=20, step=1, description='N'));
```
