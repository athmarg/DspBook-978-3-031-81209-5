```python
"""
Interactive Real Exponential Signal
Author: Athanasios I. Margaris

Purpose:
Demonstrate the behavior of real exponential continuous-time signals
defined by:

    x(t) = C * e^(alpha * t)

Features:
- Interactive sliders for parameters C and alpha
- Visualization of exponential growth and decay
- Demonstrates three important cases:
    alpha > 0  -> exponential growth
    alpha < 0  -> exponential decay
    alpha = 0  -> constant signal

Requirements:
- Python 3
- numpy
- matplotlib
- ipywidgets
"""

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Function to plot the exponential signal
def plot_exponential(C=1.0, alpha=1.0):

    t = np.linspace(-2, 2, 1000)
    x = C * np.exp(alpha * t)

    plt.figure(figsize=(8,4))
    plt.plot(t, x, linewidth=2)

    plt.xlabel("Time t")
    plt.ylabel("x(t)")
    plt.title(f"Real Exponential Signal: x(t) = {C:.2f} e^({alpha:.2f} t)")
    plt.grid(True)

    # dynamic y-limits for better visualization
    plt.ylim(min(x)-1, max(x)+1)

    plt.show()


# Interactive sliders
interact(plot_exponential,
         C=FloatSlider(value=1.0,
                       min=0.0,
                       max=10.0,
                       step=0.5,
                       description='C'),

         alpha=FloatSlider(value=1.0,
                           min=-5.0,
                           max=5.0,
                           step=0.2,
                           description='alpha'));
```
