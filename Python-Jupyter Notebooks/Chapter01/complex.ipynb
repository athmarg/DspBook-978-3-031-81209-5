```python
"""
Interactive Complex Exponential (Phasor Visualization)
Author: Athanasios I. Margaris

Signal:
    x(t) = C * exp(j*omega*t)

Visualization:
1. Rotating vector in the complex plane
2. Real part (cosine)
3. Imaginary part (sine)

Sliders:
- C : amplitude
- omega : angular frequency
- t : time instant
"""

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider


def complex_phasor_demo(C=1.0, omega=2.0, t=0.0):

    # time axis for signal plots
    t_axis = np.linspace(0, 2*np.pi, 1000)

    x = C*np.exp(1j*omega*t_axis)

    real_part = np.real(x)
    imag_part = np.imag(x)

    # current phasor position
    xt = C*np.exp(1j*omega*t)

    xr = np.real(xt)
    xi = np.imag(xt)

    fig, ax = plt.subplots(1,2,figsize=(12,5))

    # -------- complex plane --------
    ax[0].set_title("Complex Plane (Phasor)")
    ax[0].plot([0,xr],[0,xi],linewidth=3)
    ax[0].scatter(xr,xi)

    circle = plt.Circle((0,0),C,fill=False,linestyle='--')
    ax[0].add_patch(circle)

    ax[0].set_xlabel("Real")
    ax[0].set_ylabel("Imaginary")

    ax[0].set_xlim(-C*1.2,C*1.2)
    ax[0].set_ylim(-C*1.2,C*1.2)

    ax[0].grid(True)
    ax[0].set_aspect('equal')

    # -------- time signals --------
    ax[1].set_title("Real and Imaginary Components")

    ax[1].plot(t_axis,real_part,label="Real = C cos(ωt)")
    ax[1].plot(t_axis,imag_part,label="Imag = C sin(ωt)")

    ax[1].scatter(t,xr)
    ax[1].scatter(t,xi)

    ax[1].axvline(t,linestyle='--')

    ax[1].set_xlabel("time")
    ax[1].set_ylabel("value")

    ax[1].grid(True)
    ax[1].legend()

    plt.show()


interact(complex_phasor_demo,
         C=FloatSlider(value=1,min=0.5,max=5,step=0.5,description='C'),
         omega=FloatSlider(value=1,min=-5,max=5,step=0.5,description='omega'),
         t=FloatSlider(value=0,min=0,max=2*np.pi,step=0.05,description='time'));
```
