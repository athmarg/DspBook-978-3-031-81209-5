```python
# Step-function approximation of a continuous-time signal
# Demonstrates how a signal can be represented as a sum of shifted step functions

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# time axis
t = np.linspace(-5,5,2000)

# original continuous signal
def x(t):
    return np.sin(t) * np.exp(-0.2*t**2)

def step_approximation(N):

    plt.figure(figsize=(10,5))

    # plot original signal
    plt.plot(t, x(t), linewidth=3, label="Original signal x(t)")

    # number of segments used
    T = 10 / 100      # base segment width (max resolution)

    approx = np.zeros_like(t)

    # construct approximation using step functions
    for n in range(-N, N):

        tn = n*T
        tn1 = (n+1)*T

        xn = x(tn)

        # step difference u(t-nT) - u(t-(n+1)T)
        segment = xn * ((t >= tn) & (t < tn1))

        approx += segment

    plt.plot(t, approx, linewidth=2, label="Step approximation")

    plt.title(f"Signal approximation using {2*N} step segments")
    plt.xlabel("t")
    plt.ylabel("Amplitude")

    plt.grid(True)
    plt.legend()

    plt.show()

# interactive slider
interact(
    step_approximation,
    N=IntSlider(min=1,max=50,step=1,value=1,description="Segments")
)
```
