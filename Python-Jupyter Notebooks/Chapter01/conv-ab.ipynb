```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

t = np.linspace(-1,5,1000)

def u(t):
    return (t>=0).astype(float)

def x_signal(t,alpha):
    return np.exp(-alpha*t)*u(t)

def y_signal(t,beta):
    return np.exp(-beta*t)*u(t)

def convolution_result(t,alpha,beta):

    if abs(alpha-beta) > 1e-6:
        z = (np.exp(-alpha*t)-np.exp(-beta*t))/(beta-alpha)*u(t)

    else:
        z = t*np.exp(-beta*t)*u(t)

    return z


def plot_convolution(alpha,beta):

    x = x_signal(t,alpha)
    y = y_signal(t,beta)
    z = convolution_result(t,alpha,beta)

    fig,ax = plt.subplots(3,1,figsize=(8,8))

    ax[0].plot(t,x)
    ax[0].set_title(f"x(t)=exp(-αt)u(t),   α={alpha}")
    ax[0].grid(True)

    ax[1].plot(t,y)
    ax[1].set_title(f"y(t)=exp(-βt)u(t),   β={beta}")
    ax[1].grid(True)

    ax[2].plot(t,z,'r')
    ax[2].set_title("z(t)=x(t)*y(t)")
    ax[2].grid(True)
    ax[2].set_xlabel("t")

    plt.tight_layout()
    plt.show()


interact(
    plot_convolution,
    alpha=FloatSlider(min=-2,max=2,step=0.1,value=1,description='α',continuous_update=False),
    beta=FloatSlider(min=-2,max=2,step=0.1,value=0.5,description='β',continuous_update=False)
);
```
