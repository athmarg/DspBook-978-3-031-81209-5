```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, HBox, VBox, Layout

# Χρονικός άξονας
t = np.linspace(-2*np.pi, 2*np.pi, 1000)

def cross_correlation_demo(A, B, omega, tau_shift):
    # Στατικά reference σήματα
    x_ref = A * np.cos(omega * t)
    y_ref = B * np.sin(omega * t)
    
    # Sliding y(t - τ)
    y_shifted = B * np.sin(omega * (t - tau_shift))
    
    # Επικάλυψη για fill color
    overlap = np.minimum(np.maximum(x_ref, 0), np.maximum(y_shifted,0))
    
    # Θεωρητική cross-correlation για full t
    T = 2*np.pi/omega
    Rxy = - (A*B)/T * np.sin(omega * t)
    
    fig, axs = plt.subplots(2,2,figsize=(10,6))
    plt.subplots_adjust(hspace=0.4)
    
    # Άνω αριστερά: x(t)
    axs[0,0].plot(t, x_ref, 'b', lw=2)
    axs[0,0].set_title("x(t) = A cos(ωt)")
    axs[0,0].grid(True)
    axs[0,0].set_ylim(-1.2*max(A,B), 1.2*max(A,B))
    
    # Άνω δεξιά: y(t)
    axs[0,1].plot(t, y_ref, 'g', lw=2)
    axs[0,1].set_title("y(t) = B sin(ωt)")
    axs[0,1].grid(True)
    axs[0,1].set_ylim(-1.2*max(A,B), 1.2*max(A,B))
    
    # Κάτω αριστερά: Sliding y(t-τ) with fill for overlap
    axs[1,0].plot(t, x_ref, 'b', lw=2)           # x(t)
    axs[1,0].plot(t, y_shifted, 'g', lw=2)      # sliding y(t-τ)
    axs[1,0].fill_between(t, 0, overlap, color='orange', alpha=0.5)
    axs[1,0].set_title(f"Sliding y(t-τ) with τ = {tau_shift:.2f}")
    axs[1,0].set_ylim(-1.2*max(A,B), 1.2*max(A,B))
    axs[1,0].grid(True)
    
    # Κάτω δεξιά: Cross-correlation
    axs[1,1].plot(t, Rxy, 'm', lw=2)
    axs[1,1].fill_between(t, 0, Rxy, color='orange', alpha=0.5)
    axs[1,1].set_title("Cross-correlation Rxy(t)")
    axs[1,1].set_ylim(-1.2*max(A,B), 1.2*max(A,B))
    axs[1,1].grid(True)
    
    plt.show()

# Sliders
A_slider = FloatSlider(value=1, min=0.5, max=3, step=0.1, description='A', layout=Layout(width='45%'))
B_slider = FloatSlider(value=1, min=0.5, max=3, step=0.1, description='B', layout=Layout(width='45%'))
omega_slider = FloatSlider(value=1, min=0.1, max=5, step=0.1, description='ω', layout=Layout(width='45%'))
tau_slider = FloatSlider(value=0, min=-2*np.pi, max=2*np.pi, step=0.05, description='τ', layout=Layout(width='90%'))

slider_box1 = HBox([A_slider, B_slider, omega_slider])
all_sliders = VBox([slider_box1, tau_slider])

interact(cross_correlation_demo,
         A=A_slider,
         B=B_slider,
         omega=omega_slider,
         tau_shift=tau_slider)
```
