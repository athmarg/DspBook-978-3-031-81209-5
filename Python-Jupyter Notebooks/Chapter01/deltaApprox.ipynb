```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider

# Define the signal x(t)
def x(t):
    return np.sin(2*np.pi*t) + 0.5*np.cos(4*np.pi*t)

# Approximate delta function (rectangular pulse of width Delta)
def delta_approx(t, center, Delta):
    y = np.zeros_like(t)
    y[(t >= center - Delta/2) & (t <= center + Delta/2)] = 1/Delta
    return y

# Function to sum shifted delta approximations
def plot_delta_sum(N=1, Delta=0.05):
    t = np.linspace(-1, 1, 1000)
    approx = np.zeros_like(t)
    
    for n in range(-N, N+1):
        center = n * Delta
        approx += x(center) * delta_approx(t, center, Delta) * Delta  # scale by Delta for area
        
    plt.figure(figsize=(8,4))
    plt.plot(t, x(t), 'k', label='Original signal x(t)')
    plt.plot(t, approx, 'b', label=r'$\tilde{x}(t)$ approx with delta pulses')
    plt.fill_between(t, 0, approx, color='blue', alpha=0.2)
    plt.title('Approximation of x(t) using time-shifted delta functions')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid(True)
    # Legend below x-axis
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
    plt.show()

# Interactive slider
interact(plot_delta_sum,
         N=IntSlider(value=1, min=1, max=100, step=1, description='Number of pulses'),
         Delta=FloatSlider(value=0.01, min=0.01, max=0.1, step=0.005, description='Delta width'));
```
